```python
import numpy as np
import pandas as pd
np.random.seed(123)
```

```python
n_A = 1000
n_B = 500

v_A = 0.1
v_B = 4

def m_A(x):
    return x**2 + 2*x

def m_B(x):
    return x**2 + 2*x

x = np.random.uniform(-1, 1, n_A)
df_A = pd.DataFrame({
    "subset": "A",
    "y": m_A(x) + np.random.normal(0, v_A, n_A),
    "X_1": x
})

x = np.random.uniform(-1, 1, n_B)
df_B = pd.DataFrame({
    "subset": "B",
    "y": m_B(x) + np.random.normal(0, v_B, n_B),
    "X_1": x
})

# ---------- Combine ----------
df = pd.concat([df_A, df_B], ignore_index=True)
```

```python
# ---------- Save compressed ----------
df.to_csv("data/simulation2.csv.xz", index=False, compression="xz")
```
