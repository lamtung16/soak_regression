```python
import pandas as pd
from soak_high_level import SOAK
```

```python
df = pd.read_csv("data/WorkersCompensation.csv.xz")
```

```python
soak_obj = SOAK(df=df, subset_col="Gender", target_col="UltimateIncurredClaimCost")
soak_obj.analyze(model_list=["featureless", "tree"], n_splits=5, n_random_seeds=5, log_target=True)
```

```python
soak_obj.visualize(subset_value='M', model="tree", metric="rmse", figsize=(12, 2.5))
```

```python
soak_obj.visualize(subset_value='F', model="featureless", metric="rmse", figsize=(12, 2.5))
```
