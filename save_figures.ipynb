```python
from soak_low_level import SOAKFold
import pandas as pd
import os
```

```python
soak = SOAKFold()
```

```python
dataset = 'synthetic'
subset_col = 'subset'
```

```python
figures_dir = f"figures/{dataset}/{subset_col}"
os.makedirs(figures_dir, exist_ok=True)
```

```python
df = pd.read_csv(f"results/{dataset}/{subset_col}.csv")
```

```python
for subset_value in list(df['subset'].unique()):
    for model in list(df['model'].unique()):
        for metric in ['rmse', 'mae']:
            fig = soak.plot_metrics(df, subset_value, model, metric)
            fig.savefig(f"{figures_dir}/{subset_value}.{model}.{metric}.png")
```
