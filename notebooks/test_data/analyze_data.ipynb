```python
import soakpy
import pandas as pd
import warnings
import os
warnings.filterwarnings("ignore", message="FigureCanvasAgg is non-interactive")
```

```python
def try_dataset(csv_path, subset_col, target_col):
    soak_obj = soakpy.SOAK(pd.read_csv(csv_path), subset_col, target_col)
    soak_obj.analyze(n_splits=10, n_random_seeds=10)
    fig = soak_obj.visualize(figsize=(12, 3), title=csv_path[:-4])
```

```python
metadata_dict = {
    "Adrian8216_remote-worker-productivity": ("Gender", "Quality_Score"),
    "BryLut_remote-worker-productivity": ("Gender", "Quality_Score"),
    "electricsheepafrica_african-malaria-dataset": ("sex", "malaria_probability_score"),
    "electricsheepafrica_african-malnutrition-dataset": ("sex", "platelet_count"),
    "electricsheepafrica_bank-account-ownership-africa": ("gender", "account_balance_usd"),
    "electricsheepafrica_brucellosis-human-animal": ("sex", "economic_loss_usd"),
    "electricsheepafrica_climate-migration-mental-health": ("sex", "wellbeing_score"),
    "electricsheepafrica_heat-related-illness": ("sex", "work_hours_lost"),
    "electricsheepafrica_informal-sector-health-financing": ("sex", "trust_in_insurance"),
    "eliel2003_student-burnout-analysis2026": ("gender", "dropout_risk"),
    "mnemoraorg_calorie-burnt-15k": ("Gender","Calories"),
    "sadsaddfs_digital-lifestyle-benchmark-dataset": ("gender", "sleep_quality"),
    "tarekmasryo_cancer-risk-factors-data": ("Gender","Overall_Risk_Score"),
}
```

```python
for directory, (subset_col, target_col) in metadata_dict.items():
    for file_name in os.listdir(directory):
        if file_name.endswith(".csv"):
            csv_path = os.path.join(directory, file_name)
            try_dataset(csv_path, subset_col, target_col)
```
