```python
import soakpy
import pandas as pd
```

```python
df = pd.read_csv("https://huggingface.co/datasets/electricsheepafrica/technical-vocational-training-outcomes/raw/main/technical_vocational_training_outcomes_high_burden.csv")
soak_obj = soakpy.SOAK(df=df, subset_col="sex", target_col="monthly_salary_usd")
soak_obj.analyze(model_list=["tree"], n_splits=10, n_random_seeds=10)
fig = soak_obj.visualize(subset_value='male', model="tree", metric="rmse", figsize=(12, 2.5))
```
