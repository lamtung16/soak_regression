```python
import soakpy
import pandas as pd
import warnings
warnings.filterwarnings("ignore", message="FigureCanvasAgg is non-interactive")
```

```python
def try_dataset(csv_path, subset_col, target_col):
    soak_obj = soakpy.SOAK(pd.read_csv(csv_path), subset_col, target_col)
    soak_obj.analyze(n_splits=10, n_random_seeds=10)
    fig = soak_obj.visualize(figsize=(12, 2.5))
```

```python
for income in ["high", "low", "moderate"]:
    try_dataset(f"https://huggingface.co/datasets/electricsheepafrica/technical-vocational-training-outcomes/raw/main/technical_vocational_training_outcomes_{income}_burden.csv", "sex", "monthly_salary_usd")
    try_dataset(f"https://huggingface.co/datasets/electricsheepafrica/agricultural-credit-access-africa/raw/main/agricultural_credit_access_africa_{income}_burden.csv", "farmer_gender", "annual_income_usd")
```
