```python
import soakpy
import pandas as pd
import warnings
warnings.filterwarnings("ignore", message="FigureCanvasAgg is non-interactive")
```

```python
df = pd.read_csv("iid_easy.csv")
soak_obj = soakpy.SOAK(df, "Subset", "y")
soak_obj.analyze(model_list=["tree"],n_splits=10, n_random_seeds=10)
fig = soak_obj.visualize(figsize=(12, 2.5))
```
