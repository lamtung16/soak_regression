```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
pairs = [
    ("WorkersCompensation", "UltimateIncurredClaimCost"),
    ("avocado_sales", "AveragePrice"),
    ("CPMP_2015_regression", "runtime"),
]
```

```python
for dataset, target_col in pairs:
    data = np.genfromtxt(f"data/{dataset}.csv.xz", delimiter=",", dtype=None, names=True, encoding=None)
    y = data[target_col]

    fig, axes = plt.subplots(1, 2, figsize=(12, 4))

    # ---- Original scale ----
    axes[0].hist(y, bins=300)
    axes[0].set_title(f"{dataset}: {target_col} (original)")
    axes[0].set_ylabel("Count")

    # ---- Log scale ----
    axes[1].hist(np.log(y + 1), bins=300)
    axes[1].set_title(f"{dataset}: {target_col} (log scale)")

    plt.tight_layout()
    plt.show()
```
