```python
import numpy as np
from soak_low_level import SOAK

# --- synthetic data ---
X = np.arange(9).reshape(-1, 1)
y = X.ravel()
subset_vec = np.array(['even' if x % 2 == 0 else 'odd' for x in X.ravel()])

# --- Initialize soak object ---
soak_obj = SOAK(n_splits=2)
```

```python
for subset_value, category, X_train, y_train, X_test, y_test in soak_obj.split(X, y, subset_vec):
    print(f"subset: {subset_value:6s} --- category: {category:6s}")
    print(f"X_test: {X_test.ravel()}  --- y_test: {y_test}")
    print(f"X_train:", X_train.ravel(), " --- y_train:", y_train)
    print("-"*50)
```

```python
for seed in np.random.randint(1, 1000, 2):
    print("*"*50)
    X_down, y_down, subset_vec_down = soak_obj.downsample_majority(X, y, subset_vec, seed)
    for subset_value, category, X_train, y_train, X_test, y_test in soak_obj.split(X_down, y_down, subset_vec_down):
        print(f"subset: {subset_value:6s} --- category: {category:6s}")
        print(f"X_test: {X_test.ravel()}  --- y_test: {y_test}")
        print(f"X_train:", X_train.ravel(), " --- y_train:", y_train)
        print("-"*50)
```
